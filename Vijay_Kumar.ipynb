# Lending Club Case Study

## 1. Business Understanding : 
As part of the case study, we are assuming the role of a data analyst working for a consumer finance company. The company specialises in lending different type of loans to urban consumers. For every application the company faces two major risks : 
- If the customer is likely to repay the loan and gets rejected, this results in business loss to company. 
- If the customer is not likely to repay the loan and gets approved, this results in financial loss to company.

## 2. Objective

As part of the EDA process, we need to identify the factors which impact the customers ability to repay the loans. Understanding the factors impacting the loan default the company can effectively isolate risky loans and decide how to proceed with these. Once the risky applicants have been identified, the comapny can either decide to cut down on these loans or increase the interest rate and duration, this will reduce their risk exposure and help effectively manage their portfolio. 

## 3. Dataset 
The data in loan.csv contains details of the pas loan applications, and whether these were loans were repayed or not. The dataset contains
- Consumer Attributes 
- Loan Attributes

**NOTE:** Only the loans which were approved are part of the dataset, loans which were rejected have not been included.
